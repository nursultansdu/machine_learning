```
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import imutils
import zipfile
```

**Load image**

```
img = cv2.imread('/content/gdrive/My Drive/got.jpg',0)
plt.imshow(img)
```

```
#Resizing image
height, width = img.shape[:2]
print('height =', height)
print('width =', width)
img_resize = cv2.resize(img, (200, 80))
plt.imshow(img_resize)
```

```
#Rotation of image
#Calculating the center
center = (width / 2, height / 2)

#Defining angles of rotation and scaling (no scaling)
angle90 = 90
angle180 = 180
angle270 = 270
scale = 1.0

#Angle of 90 degree
matrix90 = cv2.getRotationMatrix2D(center, angle90, scale)
rotation90 = cv2.warpAffine(img, matrix90, (width, height))
plt.imshow(rotation90)
```

```
#Angle of 180 degree
matrix180 = cv2.getRotationMatrix2D(center, angle180, scale)
rotation180 = cv2.warpAffine(img, matrix180, (width, height))
plt.imshow(rotation180)
```

```
#Angle of 270 degree
matrix270 = cv2.getRotationMatrix2D(center, angle270, scale)
rotation270 = cv2.warpAffine(img, matrix270, (width, height))
plt.imshow(rotation270)
```

```
#Transformation of image
img = cv2.imread('/content/gdrive/My Drive/got.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
